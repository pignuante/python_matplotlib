```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(-5,5, 2048)

y = 0.25 * (x + 4.) * ( x + 1.) * (x - 2.)
```

## 그래프에 제목을 달기

```python
plt.title(' A polymomial ')
plt.plot(x,y, c='b')

plt.show()
```

## 4 개의 그래프를 만들고 각각 제목을 달기

```python
plt.subplot(221)
plt.title(' A polymomial ')
plt.plot(x,y, c='b')

plt.subplot(222)
plt.title(' A polymomial ')
plt.plot(x,y, c='b')

plt.subplot(223)
plt.title(' A polymomial ')
plt.plot(x,y, c='b')

plt.subplot(224)
plt.title(' A polymomial ')
plt.plot(x,y, c='b')

plt.show()
```

## 동시에 여러 개의 axes를 만들고 내부적으로 제목을 달기

```python
fig, axes4 = plt.subplots(2,2)
axes4[0,0].set_title(' A polymomial ')
axes4[0,0].plot(x,y, c='b')


axes4[0,1].set_title(' A polymomial ')
axes4[0,1].plot(x,y, c='b')

axes4[1,0].set_title(' A polymomial ')
axes4[1,0].plot(x,y, c='b')

axes4[1,1].set_title(' A polymomial ')
axes4[1,1].plot(x,y, c='b')

plt.show()
```

## Latex 표기법

    문장이 시작과 끝을  $로 포함해서 라텍스 표기법을 지정한다

$f(x) = \frac{1} {4} $

```python
plt.title(r'$f(x)= \frac{1} {4}$')
plt.plot(x,y, c='b')

plt.show()
```

$ f(x) = \frac{1}{4}{(x+4)} {(x+1)} {(x-2)}$

```python
plt.title(r'$ f(x) = \frac{1}{4}{(x+4)} {(x+1)} {(x-2)}$')
plt.plot(x,y, c='b')

plt.show()
```
