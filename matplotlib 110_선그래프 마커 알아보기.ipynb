# Marker 처리하기

### circle  marker

```python
# Make an array of x values
x = [1, 2, 3, 4, 5]
# Make an array of y values for each x value
y = [1, 4, 9, 16, 25]
# use pylab to plot x and y as red circles
plt.plot(x, y, 'ro')
```

### square marker

```python
plt.plot(x, y, 'rs')
```

### pentagon marker

```python
plt.plot(x, y, 'rp')
```

###  star marker

```python
plt.plot(x, y, 'r*')
```

### hexagon1 marke
