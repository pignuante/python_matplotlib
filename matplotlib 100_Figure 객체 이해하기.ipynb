### 현재의 figure는 gcf 함수로 가져올 수 있다.

```python
f1 = plt.figure(1)
plt.plot([1, 2, 3, 4], 'ro:')
f2 = plt.gcf()
print(f1, id(f1))
print(f2, id(f2))
plt.show()
```

```python
x = np.linspace(0, 2*np.pi)
y1 = np.sin(x);
y2 = 0.01 * np.cos(x);
```

## figure별로 그래프 그리기

```python
plt.figure()
plt.plot(x,y1)
plt.legend(['y1'])

plt.show()
```

```python
plt.figure()
plt.plot(x,y2)
plt.legend(['y2'])
plt.show()
```

```python
plt.figure()
plt.plot(x, y1, x, 100 * y2)
plt.legend(['y1', '100*y2'])
plt.show()
```

## 이중 y 축 플롯

    2 개의 분리 된 Y 축을 사용하면 스케일링 문제를 해결할 수 있습니다. 
    각 y 축은 데이터에 대해 색으로 구분됩니다. 흑백으로 인쇄 할 때이 그래프를 읽는 것이 어려울 수 있습니다.

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x, y1)
ax1.set_ylabel('y1')

ax2 = ax1.twinx()
ax2.plot(x, y2, 'r-')
ax2.set_ylabel('y2', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

print(ax1)
print(ax2)
print(ax1 is ax2)

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(x, y1)
ax1.set_xlabel('x1')

ax2 = ax1.twiny()
ax2.plot(x**2, y1, 'r-')
ax2.set_xlabel('x2', color='r')
for tl in ax2.get_xticklabels():
    tl.set_color('r')

print(ax1)
print(ax2)
print(ax1 is ax2)

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
plt.plot(x, y1)
plt.xlabel('x1')

ax2 = ax1.twiny()
plt.plot(x**2, y1, 'r-')
plt.xlabel('x2', color='r')

for tl in ax2.get_xticklabels():
    tl.set_color('r')

print(ax1)
print(ax2)
print(ax1 is ax2)

plt.show()
```

```python
# Create a new figure of size 8x6 points, using 100 dots per inch
plt.figure(figsize=(8,6), dpi=80)

ccc = plt.gcf()

# Create a new subplot from a grid of 1x1
plt.subplot(111)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Plot cosine using blue color with a continuous line of width 1 (pixels)
plt.plot(X, C, label=" x c axis ", color="blue", marker=linewidth=1.0, linestyle="-")

# Plot sine using green color with a continuous line of width 1 (pixels)
plt.plot(X, S, label=' x s axis', color="green", linewidth=1.0, linestyle="-")

# Set x limits
plt.xlim(-4.0,4.0)

# Set x ticks
plt.xticks(np.linspace(-4,4,9,endpoint=True))

# Set y limits
plt.ylim(-1.0,1.0)

# Set y ticks
plt.yticks(np.linspace(-1,1,5,endpoint=True))

# Save figure using 72 dots per inch
# savefig("../figures/exercice_2.png",dpi=72)

# Show result on screen
plt.show()
```
