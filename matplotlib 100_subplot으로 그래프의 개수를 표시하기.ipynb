## 그래프를 넣기

```python
fig = plt.figure(figsize=(10,12))
plt.subplot(411)
plt.plot(np.sqrt(a))

plt.subplot(423)
plt.plot(np.sin(a))

plt.subplot(424)
plt.plot(np.cos(a))

plt.subplot(425)
plt.plot(a**2)

plt.subplot(426)
plt.plot(np.exp(a))

plt.subplot(414)
plt.plot(np.tan(a))

plt.show()
```

## 중간 그래프를 통합하기 

### subplots_adjust  함수를 이용해서 그래프에 위치 정의

```python
help(plt.subplots_adjust)
```

```python
import  matplotlib.pyplot as plt
X = [  (4,2,1),(4,2,2), (4,2,3), (4,2,5), (4,2,(4,6)), (4,1,4)]
plt.subplots_adjust(bottom=0, left=0, top = 0.975, right=1)
for nrows, ncols, plot_number in X:
    plt.subplot(nrows, ncols, plot_number)
    plt.xticks([])
    plt.yticks([])
plt.show()
```
