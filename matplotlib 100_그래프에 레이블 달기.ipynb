## 제목 달기

```python
# Make an array of x values
x = [1, 2, 3, 4, 5]
# Make an array of y values for each x value
y = [1, 4, 9, 16, 25]
# use pylab to plot x and y
plt.plot(x, y)
# give plot a title
plt.title('Plot of y vs. x')

# show the plot on the screen
plt.show()
```

## 좌표의 레이블 달기

```python

plt.plot(x, y)
# make axis labels
plt.xlabel('x axis')
plt.ylabel('y axis')

# show the plot on the screen
plt.show()
```

## 좌표 눈금 조정하기

```python

plt.plot(x, y)
# set axis limits
plt.xlim(0.0, 7.0)
plt.ylim(0.0, 30.)

# show the plot on the screen
plt.show()
```

## 레이블에 색깔 입히기

```python
import matplotlib.pyplot as plt
import numpy as np


mpl_fig = plt.figure()
ax = mpl_fig.add_subplot(111)

ax.plot(range(10))
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

ax.xaxis.label.set_color('red')
ax.yaxis.label.set_color('green')

plt.show()
```
