# 명기적으로 그래프를 처리하기

    AxesSubplot으로 처리하면 명기적으로 그래프를 넣어서 처리한다.
    
    함수를 이용해서 처리하면 암묵적으로 그래프를 매칭시킨다.
    

## 선그래프를 axessubolot에 직접 넣기

```python
x = np.arange(8)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

### gca로 현재 axessubplot 객체를 가져와서 메소드로 그래프를 그리기

```python
ax = plt.gca()
ax.plot(x,y,label="sin")
ax.set_title(" ax sin ")
ax.legend()
plt.show()
```

```python
type(ax)
```

```python
ax.get_children()
```

## 함수를 이용해서 처리하기

```python
plt.plot(x,y,label='sin')
plt.legend()
plt.title(" func sin ")
plt.show()
```

## 범례를 latex 표기법으로 처리하기

### 메소드를 이용해서 처리하기

```python
x = np.arange(10)

fig = plt.figure()
ax = plt.subplot(111)

for i in range(5):
    ax.plot(x, i * x, label='$y = %ix$' % i)

ax.legend(bbox_to_anchor=(1.1, 1.05))

plt.show()
```

### 함수를 이용해서 처리하기

```python
x = np.arange(10)

fig = plt.figure()
plt.subplot(111)

for i in range(5):
    plt.plot(x, i * x, label='$y = %ix$' % i)

plt.legend(bbox_to_anchor=(1.1, 1.05))

plt.show()
```
