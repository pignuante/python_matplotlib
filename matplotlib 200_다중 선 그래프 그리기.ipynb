## 특정 함수 내부에 plot 함수를 정의해서 그리기

```python
def plot_slope(x,y) :
    xs = x[1:] - x[:-1]
    ys = y[1:] - y[:-1]
    plt.plot(x[1:],ys/xs)
```

```python
xx = np.linspace(-3,3,100)
```

```python
yy = np.exp(-xx **2)
```

```python
plt.plot(xx,yy)
plot_slope(xx,yy)
plt.show()
```

### 범례의 위치를 문자열로만 정의

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), label='Cosine')
ax.plot(x, np.arctan(x), label='Inverse tan')
lgd = ax.legend(loc=9)
ax.grid('on')

plt.show()
```

### 문자열과 bbox_to_anchor를 지정해서 처리

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), label='Cosine')
ax.plot(x, np.arctan(x), label='Inverse tan')
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5,1))
ax.grid('on')
```

### bbox_to_anchor 내의 음수값을 배정하면 그래프와 겹치는 부분을 없앨 수 있다.

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
fig = plt.figure(1)
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), label='Cosine')
ax.plot(x, np.arctan(x), label='Inverse tan')
handles, labels = ax.get_legend_handles_labels()
lgd = ax.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5,-0.1))
ax.grid('on')
```
