```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

## 선그래프를 axessubolot에 직접 넣기

```python
x = np.arange(8)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

### gca로 현재 axessubplot 객체를 가져와서 메소드로 그래프를 그리기

```python
ax = plt.gca()
ax.plot(x,y,label="sin")
ax.set_title(" ax sin ")
ax.legend()
plt.show()
```

```python
type(ax)
```

```python
ax.get_children()
```

## 함수를 이용해서 처리하기

```python
plt.plot(x,y,label='sin')
plt.legend()
plt.title(" func sin ")
plt.show()
```
