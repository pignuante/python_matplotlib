```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Make x, y arrays for each graph
x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
x2 = [1, 2, 4, 6, 8]
y2 = [2, 4, 8, 12, 16]
```

```python
# use pylab to plot x and y : Give your plots names
plot1 = plt.plot(x1, y1, 'r')
plot2 = plt.plot(x2, y2, 'go')
# give plot a title
plt.title('Plot of y vs. x')
# make axis labels
plt.xlabel('x axis')
plt.ylabel('y axis')
# set axis limits
plt.xlim(0.0, 9.0)
plt.ylim(0.0, 30.)

# show the plot on the screen
plt.show()
```

## 범례에 대한 세부적인 정리


     bbox_to_anchor 키워드는 수동 범례 배치에 대한 뛰어난 제어 기능을 제공합니다. 
     
     예를 들어, 축의 범례를 축의 코너 대신 그림의 오른쪽 상단에 배치하려면 모서리의 위치와 해당 위치의 좌표계를 지정하기 만하면됩니다.

```python
plt.subplot(211)
plt.plot([1,2,3], label="test1")
plt.plot([3,2,1], label="test2")
# Place a legend above this subplot, expanding itself to
# fully use the given bounding box.
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=2, mode="expand", borderaxespad=0.)

plt.subplot(223)
plt.plot([1,2,3], label="test1")
plt.plot([3,2,1], label="test2")
# Place a legend to the right of this smaller subplot.
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

## 플랏 함수에 레이블 달면 범례처리

```python
import matplotlib.pyplot as plt

line1, = plt.plot([1,2,3], label="Line 1", linestyle='--')
line2, = plt.plot([3,2,1], label="Line 2", linewidth=4)

plt.legend()

plt.show()
```

## 범례를 분리해서 표시하기

        'upper right'     1
        'lower right'     4

### 범례가 하나만 출력된다.

```python
import matplotlib.pyplot as plt

line1, = plt.plot([1,2,3], label="Line 1", linestyle='--')
line2, = plt.plot([3,2,1], label="Line 2", linewidth=4)

# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc=1)

# Create another legend for the second line.
second_legend = plt.legend(handles=[line2], loc=4)

plt.show()
```

### 범례를 분할하려면 먼저 범례를 별도의 변수에 주고 현재 그래프에 add_artist로 처리

```python
import matplotlib.pyplot as plt

line1, = plt.plot([1,2,3], label="Line 1", linestyle='--')
line2, = plt.plot([3,2,1], label="Line 2", linewidth=4)

# Create a legend for the first line.
first_legend = plt.legend(handles=[line1], loc=1)

# Add the legend manually to the current Axes.
ax = plt.gca().add_artist(first_legend)

# Create another legend for the second line.
plt.legend(handles=[line2], loc=4)

plt.show()
```
