## 확률분포에 따른 그래프 그리기

```python
def PDF(X, mu, sigma) :
    a = 1. / (sigma * np.sqrt(2. * np.pi))
    b = -1. / (2. * sigma ** 2)
    return a * np.exp(b * ( X -mu) **2)
```

```python
X = np.linspace(-5,5, 1024)
```

```python
plt.plot(X,PDF(X,0,1), color='k', linestyle='solid')
```

```python
plt.plot(X,PDF(X,0,0.5), color='k', linestyle='dashed')
```

```python
plt.plot(X,PDF(X,0,0.25), color='k', linestyle='dashdot')
```

```python
plt.plot(X,PDF(X,0,1), color='b', linestyle='dotted')
```

## 막대그래프에 선 스타일 처리하기

```python
A = np.random.random(8)
B = np.random.random(8)
```

```python
X = np.arange(8)
```

```python
plt.bar(X,A,color='0.75')
```

```python
help(plt.bar)
```

### 에지 칼라를 입혀서 처리하기

```python
plt.bar(X,A,color='0.75')
plt.bar(X,A+B, bottom=A,color='1.0',edgecolor=['0.75','0.75','0.75','0.75','0.75','0.75','0.75','0.75'],linestyle="dashed")
```
